```python
farmA=np.array([[10,12,11],[9,11,10],[12,13,14],[8,7,9]])
farmB=np.array([[20,19,21],[18,20,19],[22,23,21],[17,18,16]])
farmC=np.array([[15,16,14],[14,13,15],[16,17,18],[12,11,13]])
production_all=np.array([farmA,farmB,farmC])

# Change the dtype to float to allow adding float values
production_all = production_all.astype(np.float64)


arr=production_all
print("Shape:",arr.shape)
print("Size:",arr.size)
print("Data type:",arr.dtype)
print("Number of dimensions: ",arr.ndim)

# Calculate total energy production per farm
total_per_farm = np.sum(production_all, axis=(1, 2))
print("\nTotal energy production per farm:")
print(total_per_farm)

# Calculate mean, std, and var per farm (along axes 1 and 2)
mean_per_farm=np.mean(production_all, axis=(1, 2), keepdims=True)
print("\nMean per farm:")
print(mean_per_farm)
std_per_farm=np.std(production_all, axis=(1, 2), keepdims=True)
print("\nStandard deviation per farm:")
print(std_per_farm)
var_per_farm=np.var(production_all, axis=(1, 2), keepdims=True)
print("\nVariance per farm:")
print(var_per_farm)


hour_weights=np.array([0.5,1.0,0.75])
weighted_per_panel=np.dot(production_all,hour_weights)
print("\nWeighted per panel:")
print(weighted_per_panel)

b=0.5
# Add offset to Farm B data inside production_all using broadcasting
production_all[1] += b
print("\nProduction data with offset added to Farm B:")
print(production_all)


# Z-score normalize each farm's panel-hour readings using per-farm mean and std
# Handle potential division by zero if std is 0
normalized_production = np.where(std_per_farm != 0, (production_all - mean_per_farm) / std_per_farm, 0)
print("\nNormalized production data (z-score per farm):")
print(normalized_production)

# Apply many ufuncs (example - you can add more as needed)
print("\nApplying ufuncs (example np.exp):")
print(np.exp(normalized_production))

# You can add more ufuncs here as per the original task list
print("\nApplying ufuncs (example np.sin):")
print(np.sin(normalized_production))

# Add print statements for the rounding ufuncs as noted previously
print("\nApplying rounding ufuncs (np.floor):")
print(np.floor(production_all)) # This line was missing a print statement
print("\nApplying rounding ufuncs (np.ceil):")
print(np.ceil(production_all)) # This line was missing a print statement
print("\nApplying rounding ufuncs (np.trunc):")
print(np.trunc(production_all)) # This line was missing a print statement


high_output=production_all>20
even_hours_mask=np.array([[True,False,True],[True,False,True],[True,False,True],[True,False,True]])

print("\nLogical AND (high_output and even_hours_mask):")
print(np.logical_and(high_output[0],even_hours_mask)) # Assuming even_hours_mask applies to each farm similarly, demonstrating with farm 0

print("\nLogical OR (high_output and even_hours_mask):")
print(np.logical_or(high_output[0],even_hours_mask)) # Demonstrating with farm 0

print("\nLogical NOT (even_hours_mask):")
print(np.logical_not(even_hours_mask))

print("\nLogical XOR (high_output and even_hours_mask):")
print(np.logical_xor(high_output[0],even_hours_mask)) # Demonstrating with farm 0


print("\nGreater equal (production_all, production_all):")
print(np.greater_equal(production_all,production_all))
print("\nLess (production_all, production_all):")
print(np.less(production_all,production_all))

alerts=np.array([[1,0,1],[0,0,1],[1,1,0],[0,1,0]])

print("\nBitwise AND (alerts, alerts):")
print(np.bitwise_and(alerts,alerts))
print("\nBitwise OR (alerts, alerts):")
print(np.bitwise_or(alerts,alerts))
print("\nBitwise INVERT (alerts):")
print(np.invert(alerts))



print("\nArithmetic ufuncs (example np.add):")
print(np.add(production_all,production_all))
print("\nArithmetic ufuncs (example np.subtract):")
print(np.subtract(production_all,production_all))
print("\nArithmetic ufuncs (example np.multiply):")
print(np.multiply(production_all,production_all))
print("\nArithmetic ufuncs (example np.divide):")
print(np.divide(production_all,production_all))
print("\nArithmetic ufuncs (example np.power):")
print(np.power(production_all,production_all))
print("\nArithmetic ufuncs (example np.mod):")
print(np.mod(production_all,production_all))

print("\nAggregation over ufunc outputs (example np.sum(np.exp(normalized_production))):")
print(np.sum(np.exp(normalized_production)))
print("\nAggregation over ufunc outputs (example np.mean(np.sin(normalized_production))):")
print(np.mean(np.sin(normalized_production)))


# Broadcast-based summary table
print("\nBroadcast-based summary table:")
# Calculate panel totals per farm (sum across hours for each panel)
panel_totals_per_farm = np.sum(production_all, axis=2)
print("Panel totals per farm:")
print(panel_totals_per_farm)

# Compute needs_maintenance (True if any panel in that farm has total < 40)
needs_maintenance = np.any(panel_totals_per_farm < 40, axis=1)
print("Needs maintenance per farm:")
print(needs_maintenance)

# Final KPIs (aggregation)
overall_total=np.sum(production_all)
print("\nOverall total production:")
print(overall_total)

peak_hour_index=np.argmax(production_all)
print("\nIndex of peak hour value (flattened array):")
print(peak_hour_index)
average_per_panel=np.mean(production_all,axis=(1,2))
print("\nAverage per panel across all farms:")
print(average_per_panel)
```
