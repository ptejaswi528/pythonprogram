```python
import pandas as pd

data_1=['junior','sub junior','senior','junior','senior','sub junior','junior','senior','junior','junior','sub junior','senior','junior','senior','sub junior','junior','senior','junior']
data_2=['senior','sub junior','junior','junior','sub junior','senior','junior','senior','senior','sub junior','senior','sub junior','junior','junior','sub junior','senior','junior','senior','senior','sub junior']
data=pd.Series(data_1)
print(data)
data.value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

label_encoder = LabelEncoder()


encoded_data = label_encoder.fit_transform(data)

print("Original Data:")
print(data)
print("\nEncoded Data:")
print(encoded_data)

print("\nClasses and their encoded values:")
for i, label in enumerate(label_encoder.classes_):
    print(f"{label}: {i}")
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

one_hot_encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)


data_reshaped = data.values.reshape(-1, 1)

one_hot_encoded_data = one_hot_encoder.fit_transform(data_reshaped)

print("Original Data (first 5 entries):")
print(data.head())
print("\nOne-Hot Encoded Data (first 5 entries):")

print(one_hot_encoded_data[:5])

print("\nFeature names (categories) in order of columns:")
print(one_hot_encoder.get_feature_names_out(['category']))


one_hot_df = pd.DataFrame(one_hot_encoded_data, columns=one_hot_encoder.get_feature_names_out(['category']))
print("\nOne-Hot Encoded DataFrame (first 5 rows):")
display(one_hot_df.head())
```

```python
import pandas as pd
from google.colab import drive
from sklearn.preprocessing import OrdinalEncoder # Changed from LabelEncoder

drive.mount('/content/drive')


path='/content/drive/My Drive/loan_approved.csv'
df=pd.read_csv(path)

print(df.isnull().sum())

numeric_cols = df.select_dtypes(include=['number']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

categorical_cols = df.select_dtypes(include=['object']).columns

df[categorical_cols] = df[categorical_cols].fillna(df[categorical_cols].mode().iloc[0])



ordinal_encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1) # Use -1 for unknown values


encoded_data = ordinal_encoder.fit_transform(df[categorical_cols])

encoded_col_names = [f"{col}_encoded" for col in categorical_cols]


df[encoded_col_names] = encoded_data



encoded_columns = {col: ordinal_encoder.categories_[i] for i, col in enumerate(categorical_cols)}


print("\nDataFrame after filling NaNs and adding encoded columns (first 5 rows):")
display(df.head())
print(df.info())
df.to_csv('/content/drive/My Drive/pre_processesed.csv',index=False)

```
